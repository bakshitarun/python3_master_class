Constraints:

* 1 <= s.length <= 15

* s contains only the characters ('I', 'V', 'X', 'L', 'C', 'D', 'M').

* It is guaranteed that s is a valid roman numeral in the range [1, 3999].

```python
class Solution:
    def romanToInt( s) :
        myDict = { 'I': 1 ,'V': 5, 'X':10, 'L':50, 'C':100, 'D':500, 'M':1000}
        count =  myDict[s[0]]
        plc =  myDict[s[0]]
        for letter in range(1,len(s)):
            if myDict[s[letter]] <= plc :                
                count = count + myDict[s[letter]]
            else:
                count = count + abs( (myDict[s[letter-1]]*2) - myDict[s[letter]])
            plc = myDict[s[letter]]
        return count
    
    print('III: ' +str(romanToInt('III')))
    print('LVIII: ' +str(romanToInt('LVIII')))
    print('MCMXCIV: ' +str(romanToInt('MCMXCIV')))
    print('CCXXXIV : ' +str(romanToInt('CCXXXIV')))
    print('CXXV : ' +str(romanToInt('CXXV')))
    print('CMXIII : ' +str(romanToInt('CMXIII')))
    print('DCCLIII : ' +str(romanToInt('DCCLIII')))
    print('CLIV : ' +str(romanToInt('CLIV')))

            
```
